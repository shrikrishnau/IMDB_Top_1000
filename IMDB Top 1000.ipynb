20. Classify Movies based on Ratings [Excellent, Good, Average]

```python
def rating(IMDB_Rating):
    if IMDB_Rating>=9:
      return "Excellent"
    elif IMDB_Rating>=8:
      return "Good"
    else:
      return "Average"
```

```python
df['Rating Category']=df['IMDB_Rating'].apply(rating)
```

22. Find the unique Genres in the data

```python
list1=[]
for value in df['Genre']:
  list1.append(value.split(','))
```

```python
list2=[]
for item in list1:
  for item1 in item:
    list2.append(item1)
```

```python
uni_list=[]
for item in list2:
  if item not in uni_list:
    uni_list.append(item)
```
